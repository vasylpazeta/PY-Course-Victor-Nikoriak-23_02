# Документація: Словники (Dict) у Python

Словник (`dict`) у Python — це фундаментальна структура даних, яка зберігає довільну кількість об'єктів у вигляді пар «ключ-значення». Словники часто називають асоціативними масивами або хеш-таблицями.

---

## 1. Як працюють ключі та хеш-таблиці



Внутрішньо словники в Python реалізовані як **хеш-таблиці**. Це спеціальна структура даних, яка забезпечує неймовірно швидкий пошук, додавання та видалення елементів (в середньому за константний час `O(1)`).
___
### Механізм роботи (Хешування):
1. Коли ви додаєте пару «ключ-значення», Python пропускає ключ через спеціальну хеш-функцію.
2. Ця функція перетворює ключ на унікальне ціле число (хеш-значення).
3. Отримане число використовується як індекс (адреса "кошика" у пам'яті), куди поміщається значення.
4. Коли ви шукаєте значення за ключем, Python миттєво обчислює хеш ключа і відразу "перестрибує" до потрібного місця в пам'яті, замість того, щоб перебирати всі елементи по черзі.
____
### ⚠️ Головне правило для ключів:
Оскільки місце зберігання значення жорстко прив'язане до хешу ключа, **ключі обов'язково мають бути незмінними (immutable) та хешованими об'єктами**.

* **✅ Дозволено:** рядки (`str`), числа (`int`, `float`), логічні значення (`bool`), а також кортежі (`tuple`), якщо вони містять лише незмінні об'єкти.
* **❌ Заборонено:** списки (`list`), множини (`set`) та інші словники. Оскільки список може змінюватися, Python не може згенерувати для нього стабільний хеш. Якщо список зміниться після того, як став ключем, словник більше ніколи не знайде відповідне значення.

## 2. Для чого найчастіше використовуються словники?

* **Моделювання об'єктів та структур даних (Записи):** Словники ідеально підходять для представлення об'єктів з іменованими полями (як структури в C або об'єкти в JSON), де ключі — це назви полів, а значення — їхній вміст (наприклад, інформація про людину: `{'name': 'Bob', 'age': 40, 'job': 'dev'}`).
____
* **Створення пошукових довідників:** Словник працює як реальна телефонна книга. Замість того, щоб читати книгу від кірки до кірки, ви миттєво знаходите номер за ім'ям. Або, наприклад, система DNS, яка пов'язує доменне ім'я сайту з IP-адресою.
____
* **Кешування (Мемоізація):** Словники використовують для збереження результатів важких обчислень або запитів до сервера. Перед виконанням роботи програма перевіряє, чи є результат у словнику-кеші.
____
* **Підрахунок частоти:** Дуже поширена задача — підрахунок кількості слів у тексті або відвідувань сторінок, де ключем є слово/подія, а значенням — лічильник.

## 3. Документація по методах словника

Створити словник можна за допомогою фігурних дужок `{}` або функції `dict()`.

```
user1 = {'name': 'Ivan', 'age': 30}
# Або
user2 = dict(name='Ivan', age=30)

print("User 1:", user1)
print("User 2:", user2)
```

### 📖 Читання та пошук даних

**`get(key, [default])`**
* **Опис:** Повертає значення за заданим ключем. Якщо ключа немає, повертає `None` (або вказане `default` значення), на відміну від прямого звернення `d[key]`, яке викликає помилку `KeyError`.

**`keys()`**
* **Опис:** Повертає динамічне представлення (view) всіх ключів словника.

**`values()`**
* **Опис:** Повертає динамічне представлення всіх значень словника.

**`items()`**
* **Опис:** Повертає пари "ключ-значення" у вигляді кортежів `(key, value)`. Це найпопулярніший метод для перебору словника в циклі `for`.

```
user = {'name': 'Ivan', 'age': 30}

# get()
print("get('age'):", user.get('age', 0))
print("get('salary'):", user.get('salary', 0)) # Ключа немає, поверне 0

# keys() та values()
print("Keys:", list(user.keys()))
print("Values:", list(user.values()))

# items()
print("\nПеребір через items():")
for key, value in user.items():
    print(f"{key}: {value}")
```

### ✍️ Додавання та зміна даних

**`update(other_dict)`**
* **Опис:** Оновлює поточний словник даними з іншого словника (або ітерованого об'єкта з парами ключ-значення). Якщо ключ вже існує, його значення перезаписується; якщо ні — створюється новий.

**`setdefault(key, [default])`**
* **Опис:** Якщо ключ є у словнику, повертає його значення. Якщо ключа немає — вставляє ключ із заданим значенням `default` і повертає його.

```
# update()
d = {'a': 1, 'b': 2}
d.update({'b': 3, 'c': 4})
print("Після update:", d) # {'a': 1, 'b': 3, 'c': 4}

# setdefault()
d2 = {'a': 1}
val_b = d2.setdefault('b', 2) # Поверне 2 і додасть 'b': 2 у словник
val_a = d2.setdefault('a', 99) # Поверне 1, словник не зміниться

print("\nПісля setdefault:", d2)
print(f"Повернуті значення: val_b={val_b}, val_a={val_a}")
```

### 🗑️ Видалення даних

**`pop(key, [default])`**
* **Опис:** Видаляє пару за вказаним ключем і повертає її значення. Якщо ключа немає і вказано `default`, повертає `default`, інакше генерує помилку `KeyError`.

**`popitem()`**
* **Опис:** Видаляє і повертає останню додану пару ключ-значення у вигляді кортежу (починаючи з Python 3.7 словники зберігають порядок додавання). Зручно для використання словника як стеку (LIFO).

**`clear()`**
* **Опис:** Повністю очищає словник, видаляючи всі його елементи.

> *Примітка:* для видалення конкретного елемента без повернення його значення також часто використовують системний оператор `del`, наприклад `del d['key']`.

```
# pop()
d = {'a': 1, 'b': 2}
val = d.pop('a') 
print("Після pop('a'):", d, "| Видалене значення:", val)

# popitem()
d_stack = {'x': 10, 'y': 20}
k, v = d_stack.popitem()
print("Після popitem():", d_stack, "| Видалена пара:", k, v)

# clear() та del
d_stack.clear()
print("Після clear():", d_stack)

del_dict = {'key1': 'val1', 'key2': 'val2'}
del del_dict['key1']
print("Після del:", del_dict)
```
