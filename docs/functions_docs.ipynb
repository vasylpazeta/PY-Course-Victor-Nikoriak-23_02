# Документація: Функції та Функціональне програмування в Python

## 1. Базові концепції функцій



**Функція** — це іменований блок коду, призначений для вирішення однієї конкретної задачі. Використання функцій дозволяє уникнути дублювання коду (принцип DRY — Don't Repeat Yourself), робить програму більш читабельною, полегшує її тестування та супровід.

### Створення та виклик
У Python функції визначаються за допомогою ключового слова `def`, за яким слідує ім'я функції, круглі дужки з параметрами та двокрапка. Тіло функції виділяється відступами. Функція не виконується в момент її визначення; код усередині неї запускається лише тоді, коли функція викликається.

### Повернення значень (`return`)
Функції можуть обчислювати результат і повертати його в основну програму за допомогою інструкції `return`. Якщо функція не містить `return`, вона за замовчуванням повертає спеціальне значення `None`. Також Python дозволяє повертати одразу декілька значень (вони автоматично пакуються в кортеж).

### Документування (Docstrings)
Гарним тоном є використання рядків документації (docstrings) — багаторядкових коментарів у потрійних лапках `"""..."""`, які розміщуються одразу після визначення `def`. Цей текст автоматично зберігається у спеціальному атрибуті функції `__doc__`. Прочитати його можна за допомогою вбудованої функції `help(ім_я_функції)`.

```
def greet_user(name):
    """
    Виводить просте вітання для вказаного користувача.
    """
    return f"Hello, {name}!"

# Виклик функції
message = greet_user("Ivan")
print(message)

# Перегляд документації
help(greet_user)
```

## 2. Робота з аргументами та параметрами



Python має надзвичайно гнучкий механізм передачі аргументів.

* **Позиційні та іменовані аргументи:** Аргументи можна передавати в тому порядку, в якому вони визначені (позиційні), або вказуючи їхні імена (іменовані), що робить порядок неважливим.
* **Значення за замовчуванням:** Параметрам можна задати значення за замовчуванням (наприклад, `def func(a, b=2):`). Тоді при виклику цей аргумент можна пропустити. 
  > ⚠️ **Важливо:** ніколи не використовуйте змінні об'єкти (як-от порожні списки `[]`) як значення за замовчуванням, оскільки вони зберігатимуть свій стан між викликами. Замість цього використовуйте `None`.
* **Довільна кількість аргументів (`*args` та `**kwargs`):**
  * Символ `*` перед параметром (зазвичай `*args`) збирає всі додаткові позиційні аргументи в **кортеж**.
  * Символи `**` (зазвичай `**kwargs`) збирають додаткові іменовані аргументи у **словник**.

```
def my_func(a, b=10, *args, **kwargs):
    print("a:", a)
    print("b:", b)
    print("args:", args)
    print("kwargs:", kwargs)

my_func(1, 20, 30, 40, x=100, y=200)
# Результат:
# a: 1
# b: 20
# args: (30, 40)
# kwargs: {'x': 100, 'y': 200}
```

## 3. Область видимості (Scopes)

Змінні, створені всередині функції, є локальними і не існують за її межами. Якщо функція намагається знайти змінну, Python шукає її за правилом **LEGB**:
1. **L**ocal (Локальна) — всередині функції.
2. **E**nclosing (Охоплююча) — у зовнішній функції (якщо функції вкладені).
3. **G**lobal (Глобальна) — на рівні модуля/файлу.
4. **B**uilt-in (Вбудована) — базові імена Python (наприклад, `len`, `print`).

Щоб змінити глобальну змінну всередині функції, використовується ключове слово `global`. Для зміни змінної з охоплюючої (зовнішньої) функції використовується `nonlocal`.

## 4. Функціональне програмування (FP) в Python



Хоча Python не є суто функціональною мовою програмування, він містить потужний арсенал інструментів для написання коду у функціональному стилі. Основа ФП — обчислення без зміни глобального стану (використання "чистих функцій", які не мають побічних ефектів).

### Функції як об'єкти першого класу
У Python абсолютно все є об'єктами, і функції — не виняток. Функцію можна присвоїти змінній, передати як аргумент іншій функції (функції вищого порядку) або повернути як результат.

### Анонімні функції (`lambda`)
Для створення коротких, одноразових функцій використовується оператор `lambda`. Вони містять лише один вираз, результат якого автоматично повертається.

### Основні інструменти: `map`, `filter`, `reduce`
* **`map(func, iterable)`:** Застосовує функцію до кожного елемента послідовності.
* **`filter(func, iterable)`:** Повертає лише ті елементи, для яких функція повернула `True`.
* **`reduce(func, iterable)`:** Послідовно застосовує функцію до пар елементів, "згортаючи" послідовність до одного значення. Потребує імпорту з `functools`.

*(Примітка: у сучасному Python замість `map` та `filter` часто віддають перевагу генераторам списків - List Comprehensions, оскільки вони читаються легше).*

```
from functools import reduce

# Лямбда-функція
add_lambda = lambda x, y: x + y
print("Лямбда 5+3 =", add_lambda(5, 3))

nums = [1, 2, 3, 4, 5]

# Map: підносимо до квадрату
squares = list(map(lambda x: x**2, nums))
print("Map (квадрати):", squares)

# Filter: залишаємо лише парні
evens = list(filter(lambda x: x % 2 == 0, nums))
print("Filter (парні):", evens)

# Reduce: сума всіх чисел
total = reduce(lambda a, b: a + b, nums)
print("Reduce (сума):", total)
```

### Корисні модулі для ФП: `operator` та `functools`
* **Модуль `operator`:** Містить функції, що відповідають стандартним математичним операторам. Замість `lambda x, y: x + y` можна імпортувати `operator.add`.
* **`functools.partial`:** Дозволяє "зафіксувати" частину аргументів функції, створивши нову функцію з меншою кількістю параметрів (часткове застосування).

```
from functools import partial
import operator

def power(base, exp): 
    return base ** exp

# Створюємо нову функцію, де exp завжди дорівнює 2
square = partial(power, exp=2) 
print("Partial (5^2):", square(5))  # Результат: 25

# Використання operator замість lambda в reduce
total_op = reduce(operator.add, nums)
print("Reduce з operator.add:", total_op)
```

## 5. Просунуті техніки на базі функцій



* **Вкладені функції та Замикання (Closures):** Python дозволяє визначати функцію всередині іншої функції. Вкладена функція може запам'ятовувати та використовувати змінні із зовнішньої (охоплюючої) функції навіть після того, як зовнішня функція завершила свою роботу. Це явище і називається замиканням.
* **Декоратори:** Це спеціальні функції-обгортки, які приймають іншу функцію як аргумент, модифікують її поведінку і повертають нову функцію. Застосовуються за допомогою символу `@` перед оголошенням `def`. Ідеально для логування або кешування.
* **Рекурсія:** Функція може викликати саму себе. Корисно для обробки деревоподібних структур даних або математичних алгоритмів (наприклад, факторіал), проте завжди потрібно передбачати умову виходу, щоб уникнути нескінченного циклу.

```
# --- Приклад 1: Декоратор ---
def uppercase_decorator(func):
    def wrapper(*args, **kwargs):
        original_result = func(*args, **kwargs)
        return original_result.upper() # Змінюємо поведінку (робимо великими літерами)
    return wrapper

@uppercase_decorator
def say_hi(name):
    return f"Привіт, {name}"

print("Декоратор:", say_hi("Олексій")) # Результат: ПРИВІТ, ОЛЕКСІЙ


# --- Приклад 2: Рекурсія ---
def factorial(n):
    if n == 1: # Умова виходу (Base case)
        return 1
    return n * factorial(n - 1) # Рекурсивний виклик

print("\nРекурсія (5!):", factorial(5)) # Результат: 120
```
