# Документація: Кортежі (`tuple`) та Іменовані кортежі (`namedtuple`)

## 1. Базовий кортеж (`tuple`)



**Кортеж (`tuple`)** — це вбудована структура даних у Python, яка працює майже так само, як список (`list`), але з однією критичною відмінністю: кортежі є **незмінними (immutable)** послідовностями. Це означає, що після створення об'єкта-кортежу ви не можете змінити, додати або видалити його елементи.

* **Створення:** Створюється за допомогою круглих дужок `()` або функції `tuple()`. Елементи розділяються комами.

**Для чого використовується:**
1. Для зберігання даних, які не повинні змінюватися протягом роботи програми (наприклад, координати, налаштування).
2. Для повернення декількох значень із функції.
3. Як ключі для словників (`dict`), оскільки кортежі є хешованими (на відміну від списків).

```
# 1. Базове створення та спроба зміни
point = (10, 20, 30)
print("Координата X:", point[0])  # Результат: 10
# point[0] = 15  # ❌ Викличе помилку TypeError! Кортежі незмінні

# 2. Реальний приклад: Повернення декількох значень з функції
def get_screen_resolution():
    # Уявімо, що тут логіка визначення екрану
    return 1920, 1080 # Python автоматично пакує це в кортеж!

width, height = get_screen_resolution() # Розпакування кортежу
print(f"Роздільна здатність: {width}x{height}")

# 3. Кортеж як ключ для словника (зі списком так не вийде)
map_grid = {
    (0, 0): "Старт",
    (1, 2): "Скарб",
    (5, 5): "Ворог"
}
print("Що на позиції (1, 2)?", map_grid[(1, 2)])
```

### Методи кортежу:
Оскільки кортеж незмінний, він має лише два методи:
* `count(x)` — повертає кількість входжень елемента `x` у кортеж.
* `index(x)` — повертає індекс першого входження елемента `x`.

```
history = ('login', 'click', 'view', 'click', 'logout')
print("Кількість кліків:", history.count('click'))
print("Індекс першого перегляду:", history.index('view'))
```

## 2. Іменований кортеж (`namedtuple`) — "Незмінний словник"



Хоча звичайний `tuple` дуже швидкий і зручний, у нього є недолік: щоб отримати доступ до даних, потрібно пам'ятати їхні числові індекси. Якщо полів багато (2-3 і більше), це стає незручним.

Для вирішення цієї проблеми існує **іменований кортеж (`namedtuple`)**. Він дозволяє використовувати цю структуру як простий незмінний контейнер, який поєднує гнучкість повноцінних класів та зручність словників. Ця фабрика класів створює підкласи звичайного `tuple`, тому екземпляри іменованого кортежу залишаються абсолютно незмінними.

### Чому `namedtuple` схожий на словник (`dict`), але кращий?
* **Доступ за іменами:** Ви можете звертатися до значень не тільки за числовими індексами або через ітерування (що все ще підтримується), але й за зрозумілими іменами атрибутів.
* **Незмінність (Immutability):** На відміну від словника, ви не можете випадково перезаписати значення. Це гарантує цілісність даних.
* **Захист від помилок:** Використання `namedtuple` усуває необхідність писати імена полів як рядкові ключі (як у словниках), що дозволяє уникнути помилок через опечатки.
* **Конвертація в `dict`:** Будь-який `namedtuple` має вбудований метод `._asdict()`, який дозволяє миттєво сконструювати звичайний словник `dict` з його полів.

```
from collections import namedtuple

# 1. Оголошення структури (шаблону)
Car = namedtuple('Car', 'color mileage') 

# 2. Створення екземпляра
my_car = Car(color='червоний', mileage=1234)

# 3. Доступ до даних (ЯК У СЛОВНИКУ, але через крапку)
print("Колір:", my_car.color)
print("Пробіг:", my_car.mileage)

# 4. Доступ за індексом зберігається
print("Індекс 0:", my_car[0])

# 5. Конвертація у словник
car_dict = my_car._asdict()
print("Словник:", car_dict)

# Реальний приклад: Рядок з бази даних
DBRow = namedtuple('UserRow', ['id', 'email', 'role'])
user_record = DBRow(id=404, email='admin@site.com', role='superadmin')

print(f"\nКористувач {user_record.email} має роль {user_record.role}")
```

## 3. Сучасний підхід: `typing.NamedTuple`

`typing.NamedTuple` — це сучасний, вдосконалений "молодший брат" класичного `collections.namedtuple`. Починаючи з Python 3.6, він отримав підтримку зручного синтаксису класів (через PEP 526) та можливість додавати анотації типів.

### Синтаксис та базове створення
На відміну від старого підходу, де параметри передавалися рядком, `NamedTuple` використовує звичайне визначення класу. Кожне поле **обов'язково** повинно мати анотацію типу.

Завдяки цьому код стає набагато читабельнішим, а в IDE працює автодоповнення та підсвічування помилок.

```
from typing import NamedTuple

# Оголошення структури
class UserProfile(NamedTuple):
    user_id: int
    phone: str
    status: str = "Active" # Можна задати значення за замовчуванням

# Створення об'єкта
user1 = UserProfile(user_id=101, phone="+380991234567")
print("Телефон:", user1.phone)
print("Статус (за замовчуванням):", user1.status)
```

### 🔑 Ключові особливості `typing.NamedTuple`
1. **Абсолютна незмінність (Immutability):** Оскільки `NamedTuple` є підкласом звичайного `tuple`, його екземпляри не можна змінювати після створення. Це гарантує цілісність даних.
2. **Типізація лише для статичного аналізу:** Python ігнорує ці типи під час виконання (runtime). Тобто ви технічно можете передати рядок замість числа, і код не впаде. Однак інструменти перевірки типів (наприклад, mypy) відловлять цю помилку.
3. **Гібридний доступ:** Ви можете отримувати дані як за іменами атрибутів, так і за числовими індексами. Також підтримується розпакування.
4. **Економічність:** Екземпляри `NamedTuple` не використовують словник `__dict__` під капотом, тому вони споживають так само мало оперативної пам'яті, як і звичайні кортежі.

### 🛠 Додавання власних методів та вбудовані функції
Величезна перевага використання синтаксису класів полягає в тому, що ви можете додавати власні методи прямо в тіло `NamedTuple`.
Також він зберігає сумісність із класичним `namedtuple` і має вбудовані методи (починаються з підкреслення):
* `._asdict()` — повертає дані у вигляді словника.
* `._replace(**kwargs)` — створює **новий** екземпляр, замінюючи вказані поля новими значеннями.
* `._fields` — повертає кортеж рядків з іменами всіх полів.
* `.__annotations__` — словник з інформацією про типи.

```
class Coordinate(NamedTuple):
    lat: float
    lon: float
    reference: str = 'WGS84'

    # Власний метод
    def get_formatted(self) -> str:
        ns = 'N' if self.lat >= 0 else 'S'
        we = 'E' if self.lon >= 0 else 'W'
        return f"{abs(self.lat):.1f}°{ns}, {abs(self.lon):.1f}°{we}"

moscow = Coordinate(55.756, 37.617)
print("Форматований вивід:", moscow.get_formatted())

# Використання вбудованих методів
new_coord = moscow._replace(lon=40.000) # Зміна даних через створення копії
print("Нова координата:", new_coord)
print("Поля структури:", Coordinate._fields)
```

## 4. Чому варто використовувати `typing.NamedTuple` замість `dict`?

Використання `typing.NamedTuple` замість звичайних словників (`dict`) має низку суттєвих переваг, особливо під час створення записів або об'єктів для зберігання структурованих даних:

1. **Економія пам'яті:** Об'єкти `NamedTuple` реалізовані як звичайні класи Python на основі кортежів, тому вони використовують пам'ять так само ефективно, як і звичайні кортежі. Словники ж потребують значно більше місця в пам'яті через використання хеш-таблиць та необхідність зберігати імена ключів для кожного запису.
2. **Незмінність і цілісність даних:** На відміну від словників, які є змінними (можна вільно додавати, видаляти або змінювати елементи), `NamedTuple` створює незмінні об'єкти. Це гарантує цілісність даних у програмі — після створення екземпляра ви не зможете випадково перезаписати його поля.
3. **Захист від помилок (одруківок):** Під час роботи зі словниками немає механізмів, які б захищали від неправильного написання ключів. `NamedTuple` працює як жорсткий шаблон, гарантуючи використання лише правильних імен полів.
4. **Зручний доступ до полів:** Замість доступу до значень за рядковими ключами (наприклад, `user['name']`), `NamedTuple` дозволяє звертатися до полів як до атрибутів об'єкта (`user.name`), що робить код набагато чистішим і зрозумілішим.
5. **Статична перевірка типів:** Підтримка анотацій типів для полів допомагає інструментам статичного аналізу (наприклад, `mypy`) виявляти помилки ще до запуску програми.
6. **Зручний синтаксис `class` та розширюваність:** Дозволяє легко вказувати значення за замовчуванням для полів, а також додавати в структуру власні методи чи рядки документації, що недоступно для звичайних словників.

### 🎓 Висновок
Експерти рекомендують використовувати `typing.NamedTuple` як **"варіант за замовчуванням" у Python 3** для створення простих записів, структур або об'єктів перенесення даних (DTO). Якщо вам потрібна проста, ефективна за пам'яттю структура з незмінними даними, захистом від опечаток та перевіркою типів, `typing.NamedTuple` є набагато кращим вибором, ніж звичайний словник.

```
# ПРАКТИЧНЕ ПОРІВНЯННЯ: Обробка конфігурації

# ❌ Підхід зі словником (Dict)
config_dict = {
    "host": "localhost",
    "port": 8080,
    "debug": True
}
# Легко помилитися в імені ключа і програма впаде не одразу
# Легко випадково перезаписати: config_dict['port'] = 9999
print("Dict Config:", config_dict.get("host"))

# ✅ Підхід з NamedTuple
class ServerConfig(NamedTuple):
    host: str
    port: int
    debug: bool = False

config_obj = ServerConfig(host="localhost", port=8080, debug=True)
# config_obj.port = 9999 # Одразу підсвітить помилку (незмінність!)
# config_obj.hst # Одразу підсвітить помилку (захист від опечаток!)
print("NamedTuple Config:", config_obj.host)
```
