## 1. Булеві значення та квазіістинність

### Тип bool

`bool` у Python — це **підклас** цілого числа `int`:

| Значення | Числовий еквівалент |
|----------|---------------------|
| `True`   | `1`                 |
| `False`  | `0`                 |

Тому арифметика з `bool` цілком законна:

```python
True + True   # → 2
True + False  # → 1
```

---

### Truthy та Falsy об'єкти

Кожен об'єкт у Python неявно оцінюється як `True` або `False` в умовних виразах.

**Завжди False (Falsy):**

| Значення | Тип      |
|----------|----------|
| `0`      | int      |
| `0.0`    | float    |
| `""`     | str      |
| `[]`     | list     |
| `{}`     | dict     |
| `None`   | NoneType |

**Все інше** — `True` (Truthy): `"False"`, `-1`, `[0]`, `" "` тощо.

````python
# bool — підклас int
print("True + True  =", True + True)
print("True + False =", True + False)
print("isinstance(True, int) =", isinstance(True, int))

print()

# Falsy значення
falsy_values = [0, 0.0, "", [], {}, None]
print("Falsy значення:")
for v in falsy_values:
    print(f"  bool({v!r:8}) = {bool(v)}")

print()

# Truthy значення
truthy_values = [1, -1, "0", [0], " "]
print("Truthy значення:")
for v in truthy_values:
    print(f"  bool({v!r:8}) = {bool(v)}")
````

## 2. Логічні оператори та пріоритет

| Оператор | Опис                                    |
|----------|-----------------------------------------|
| `not`    | заперечення: `not True` → `False`       |
| `and`    | обидва True → True                      |
| `or`     | хоча б один True → True                 |

### Пріоритет (від вищого до нижчого)

```
not  →  and  →  or
```

Тому вираз `not x and y or z` читається як `((not x) and y) or z`.

---

### Short-circuiting (укорочена оцінка)

Python зупиняє обчислення, як тільки результат стає очевидним:

```python
True or f()    # f() НЕ викликається
False and f()  # f() НЕ викликається
```

Практичне застосування — безпечне ділення:

```python
if y != 0 and x / y > 1:   # ділення не відбудеться якщо y == 0
    ...
```

````python
# Пріоритет: not > and > or
x, y = True, False

print("x =", x, "  y =", y)
print()
print("not x              =", not x)
print("not x and y        =", not x and y)       # (not x) and y
print("not x and y or x   =", not x and y or x)  # ((not x) and y) or x

print()

# Short-circuiting
def f():
    print("  ← f() була викликана!")
    return True

print("True or f():")
result = True or f()   # f() не викликається
print("  result =", result)

print()
print("False and f():")
result = False and f() # f() не викликається
print("  result =", result)
````

## 3. None — відсутність значення

`None` — спеціальний об'єкт типу `NoneType`:

- Існує **лише один** `None` в Python (singleton)
- Означає «значення відсутнє» (аналог `null` в інших мовах)
- Функція без `return` **завжди повертає `None`**

```python
def greet():
    print("Hi")

result = greet()   # result == None
```

**Правильна перевірка:**

```python
if value is None:    # ✅ рекомендовано
    ...

if value == None:    # ⚠️ працює, але не рекомендовано
    ...
```

````python
def greet():
    print("Hi")

result = greet()

print("result        =", result)
print("type(result)  =", type(result))
print("result is None:", result is None)
print("result == 0:  ", result == 0)
print("result == '': ", result == "")

print()

# Singleton: усі None — один і той самий об'єкт
a = None
b = None
print("id(a) == id(b):", id(a) == id(b))  # True — один об'єкт
````

## 4. Умовні конструкції: if / elif / else

```python
if умова1:
    # виконується якщо умова1 True
elif умова2:
    # виконується якщо умова2 True (і умова1 False)
else:
    # виконується якщо жодна умова не True
```

**Ключове правило:** Python перевіряє умови **зверху вниз**.  
Як тільки знаходить першу `True` — виконує блок і виходить з конструкції.

```python
score = 85
if score >= 60:    # ← True! Виконується цей блок.
    print("Склав")
elif score >= 80:  # ← ніколи не перевіряється
    print("B")
```

> **Порядок умов має значення:** розміщуйте від **конкретніших** до **загальніших**.

````python
# ❌ Неправильний порядок — elif ніколи не спрацює
print("=== Неправильний порядок ===")
score =85
if score >= 60:
    print("Склав (score =", score, ")")
elif score >= 80:          # ← ніколи не досягається
    print("Оцінка B")

print()

# ✅ Правильний порядок — від строгішого до загального
print("=== Правильний порядок ===")
for score in [95, 82, 65, 40]:
    if score >= 90:
        grade = "Відмінно"
    elif score >= 80:
        grade = "Добре"
    elif score >= 60:
        grade = "Склав"
    else:
        grade = "Не склав"
    print(f"  score={score} → {grade}")
````

## 5. Цикл while

```python
while умова:
    # тіло циклу
    # ← обов'язково змінювати умову!
```

Підходить коли кількість ітерацій **наперед невідома**.

**Типова помилка — нескінченний цикл:**

```python
x = 1
while x <= 5:
    print(x)   # x ніколи не змінюється → зависання ☠️
```

**Правильно:**

```python
x = 1
while x <= 5:
    print(x)
    x += 1     # наближаємо умову до False
```

````python
# Простий while
print("Рахуємо до 5:")
x = 1
while x <= 10:
    print(x, end=" ")
    x += 1
print()

print()

# Практичний while: очікування коректного числа
# (симуляція — замість input() підставляємо список значень)
attempts = ["abc", "", "42"]   # що «вводить» користувач
attempt_index = 0

print("Симуляція введення числа:")
while True:
    user_input = attempts[attempt_index]
    attempt_index += 1
    print(f"  Введено: {user_input!r}", end=" → ")
    if user_input.isdigit():
        print("✅ Прийнято:", int(user_input))
        break
    else:
        print("❌ Не число, спробуйте ще")
````

## 6. Управління циклом: break та continue

| Оператор   | Що робить                         |
|------------|-----------------------------------|
| `break`    | повністю зупиняє цикл             |
| `continue` | пропускає поточну ітерацію        |

```python
for i in range(1, 6):
    if i == 3:
        continue  # пропустити 3
    if i == 5:
        break     # зупинитись перед 5
    print(i)

# виведе: 1 2 4
```

````python
# continue — пропустити ітерацію
print("continue (пропускаємо парні):")
for i in range(1, 8):
    if i % 2 == 0:
        continue
    print(i, end=" ")
print()

print()

# break — вийти з циклу
print("break (зупинитись на першому від'ємному):")
numbers = [3, 7, 1, -2, 8, -5]
for n in numbers:
    if n < 0:
        print(f"  Знайдено від'ємне: {n} → break")
        break
    print(f"  {n} — позитивне, продовжуємо")

print()

# break + continue разом
print("break + continue (range 1–5, пропуск 3, зупинка на 5):")
for num in range(1, 6):
    if num == 3:
        continue
    if num == 5:
        break
    print(num, end=" ")
print()
````

## 7. Потік виконання: `__name__ == "__main__"`

Уявіть, що ваш файл із кодом — це людина, яка може виступати у **двох різних ролях**:

1. **«Господар»** — коли ви запускаєте саме цей файл напряму (як самостійну програму).
2. **«Гість»** — коли інший файл підключає його до себе за допомогою команди `import`, щоб просто скористатися його функціями.

Спеціальна вбудована змінна `__name__` — це своєрідний **«бейдж»**, який інтерпретатор Python автоматично чіпляє на ваш файл, щоб файл розумів, у якій ролі він зараз перебуває.

---

### Як змінюється цей «бейдж»

- Коли ви **запускаєте файл напряму** (наприклад, `python my_program.py`), Python присвоює змінній `__name__` значення `"__main__"`.
- Коли ви **імпортуєте файл** в інший скрипт (`import my_program`), Python присвоює `__name__` значення рівне реальній назві файлу (наприклад, `"my_program"`).

---

### Що робить рядок `if __name__ == "__main__":`?

Це звичайна умова, яка буквально перевіряє:  
*«Цей файл зараз запустили як головну програму?»*

- Якщо **ТАК** — викликається `main()` або будь-який інший код під цим `if`.
- Якщо **НІ** (файл просто імпортували) — блок повністю ігнорується.

---

### Чому це найкраща практика?

У Python під час імпорту файлу інтерпретатор **одразу виконує всі інструкції** на верхньому рівні.

Якби ви не сховали запуск програми під `if __name__ == "__main__":`, то при спробі просто імпортувати ваш файл — ваш код почав би раптово виконуватися: щось друкувалося б на екран, запускалася б логіка гри або тести. Інша програма просто хотіла б «позичити» пару функцій, а натомість отримала б купу непотрібних дій.

Використовуючи цю конструкцію, ви робите свій код **універсальним**: він працює і як самостійний скрипт, і як тиха «бібліотека», яка нічого зайвого не робить при імпорті.

---

| Сценарій             | `__name__`     | Блок `if` виконується? |
|----------------------|----------------|------------------------|
| Запуск напряму       | `"__main__"`   | ✅ так                 |
| Імпорт як модуль     | назва файлу    | ❌ ні                  |

````python
# Демонстрація __name__ у поточному середовищі
print("Значення __name__:", __name__)
print()

if __name__ == "__main__":
    print("✅ Файл запущено напряму — блок виконується")
else:
    print("📦 Файл імпортовано як модуль — блок пропущено")

print()

# Типовий патерн у реальних програмах
def process_data(data):
    return [x * 2 for x in data]

def main():
    data = [1, 2, 3, 4, 5]
    result = process_data(data)
    print("Результат:", result)

# Цей блок виконується ТІЛЬКИ при прямому запуску:
if __name__ == "__main__":
    print("hi")
    main()
````
